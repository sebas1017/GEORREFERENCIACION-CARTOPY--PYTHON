# Solucion punto  3 graficar capitales en el mapa de colombia

```python
#utilizando la libreria del modulo cartopy en concreto el metodo feature (as cf) puedo traer el mapa de colombia
mapa_libreria = cf.NaturalEarthFeature(category='cultural',  
                name='admin_0_boundary_lines_land',
                scale='10m',facecolor='none')  #aqui me traigo el mapa utilizando el modulo cf de la clase feature
stt_prv = cf.NaturalEarthFeature(category='cultural', 
                name='admin_1_states_provinces_lines',
                scale='10m',facecolor='none')
fig = plt.figure(figsize=(50, 50))     #planteo el tamaño inicial del mapa   
ax = fig.add_subplot(2, 5, 10, projection=ccrs.PlateCarree())#defino que en esta ventana se usen imagenes
ax.set_extent([-65, -80, -5, 13])# planteo las medidas del mapa
ax.set_title("CAPITALES DE COLOMBIA", fontsize=15)
ax.coastlines()

#aqui leo el diccionario de ciudades y tomo el par ordenado x,y y con el metodo scatter voy graficando ese x,y 
#y  lo guardo en el objeto ax que contiene la informacion del grafico , luego en un ciclo voy agregando esos pares ordenados
# y cuando el ciclo finaliza lo grafico con plt.show()
[ax.scatter ( valor[elemento] , valor[elemento+1]) for llave , valor in ciudades.items() for elemento in range(len(valor)-1)]
#list comprehension

          
ax.add_feature(mapa_libreria,     linewidth=0.5, edgecolor='black', zorder=0)
ax.add_feature(stt_prv, linewidth=0.2, edgecolor='black', zorder=5)       
plt.show()
#PUNTO 3 GRAFICAR CADA UNO DE LAS CAPITALES EN EL MAPA MEDIANTE LAS COORDENADAS SOLICITADAS
```

# Solucion punto   1.a  , 1.b , 2.a, 2.b

```python
#lista donde almacenare todos los datos
#se limpiaron los datos del csv de  forma manual eliminando los registros que les faltaba datos
datos_cargados = []
with open("Deptos_Colombia.csv") as archivo:
    datos = csv.reader(archivo, delimiter = ";")
    for linea in datos:
        datos_cargados.append(linea)
        #cargamos los datos....
        #leemos el archivo y cada linea es una lista y estas son agregadas a datos_cargados
        #formando una estructura de datos final tipo matriz [[],[],[]]
        
#Recorro la matriz
for dato in range(len(datos_cargados)):
    #valido que la primera sublista sea la que contiene los encabezados
    #para asi agregar las columnas adicionales a los datos originales
    if dato ==0:
        #agrego las columnas que seran transformadas de habitantes a porcentajes
        datos_cargados[dato].append("Porcentaje Pobres")
        datos_cargados[dato].append("Porcentaje Vulnerables")
        datos_cargados[dato].append("Porcentaje Clase Media")
        datos_cargados[dato].append("Porcentaje Clase Alta")
        #agrego las columnas que seran transformadas de porcentajes a habitantes
        datos_cargados[dato].append("Habitantes_Edad_Trabajar")
        datos_cargados[dato].append("Habitantes_Econ_Activa")
        datos_cargados[dato].append("Habitantes_Ocupados")
        datos_cargados[dato].append("Habitantes_Desocupados")
        datos_cargados[dato].append("Poblacion_Total")

    else:
        #para las demas sublistas que ya serian los datos del csv entonces les agrego los datos correspondientes
        #a las columnas que agregue anteriormente
        total_poblacion = int(datos_cargados[dato][4])+int(datos_cargados[dato][5])#poblacion_cabecera+poblacion_resto
        #empezamos a agregar los porcentajes para cada feature adicional que se agrego anteriormente(porcentajes) 
        datos_cargados[dato].append(int(datos_cargados[dato][6]) / total_poblacion)
        datos_cargados[dato].append(int(datos_cargados[dato][7]) / total_poblacion)
        datos_cargados[dato].append(int(datos_cargados[dato][8]) / total_poblacion)
        datos_cargados[dato].append(int(datos_cargados[dato][9]) / total_poblacion)
        #aqui ya agregamos los porcentajes de las 4 caracteristicas que iban de habitantes a porcentajes
        # A continuacion agregaremos los valores de las columnas que van de porcentaje a numero de habitantes
        datos_cargados[dato].append(float(datos_cargados[dato][10]) * total_poblacion)
        datos_cargados[dato].append(float(datos_cargados[dato][11]) * total_poblacion)
        datos_cargados[dato].append(float(datos_cargados[dato][12]) * total_poblacion)
        datos_cargados[dato].append(float(datos_cargados[dato][13]) * total_poblacion)
        datos_cargados[dato].append((datos_cargados[dato][2] ,total_poblacion))

        
        
```

# Solucion del punto 4 y punto 5

```python

#Aqui declaramos las caracteristicas disponibles para cambiar el tamaño proporcional de los circulos graficados en el mapa
#en base a la caracteristica escogida se graficaran los puntos en el mapa en base a las capitales que el csv tiene
#en este caso ya no estaran todas las capitales del diccionario ciudades ya que no todas  tienen datos en el csv
print("SUPERFICIE KM2")
print("POBLACION")
print("NUMERO DE HABITANTES POBRES")
print("NUMERO DE HABITANTES VULNERABLES")
print("NUMERO DE HABITANTES CLASE MEDIA")
print("NUMERO DE HABITANTES CLASE ALTA")
print("INDICE DE DESARROLLO HUMANO")
print("PORCENTAJE HABITANTES CON EDAD TRABAJAR")
print("PORCENTAJE HABITANTES ECONOMICAMENTE ACTIVOS")
print("PORCENTAJE HABITANTES OCUPADOS")
print("PORCENTAJE HABITANTES DESOCUPADOS")
print("MUNICIPIOS")
print("0 PARA SALIR DEL PROGRAMA")

#se debe digitar en el input tal cual esta el texto para poder usar el programa
opcion = input("INGRESE LA CARACTERISTICA  PARA GRAFICAR DE FORMA PROPORCIONAL LOS PUNTOS:"+"\n")
print("ESCRIBA 0 SI DESEA SALIR")
while(opcion!="0"):
    #ciclo mientras el usuario quiera seguir viendo datos
    #con clear_output limpio la salida de la opcion anterior para graficar la nueva opcion del usuario
    clear_output(wait=True)   
    resultado_totales_caracteristica = {}  
    totales_poblacion_capital  =[]
    for elemento in datos_cargados:
        if opcion == "POBLACION":
            totales_poblacion_capital.append(elemento[24])
            color = "y"
        if opcion == "NUMERO DE HABITANTES POBRES":
            color = "b"
            totales_poblacion_capital.append((elemento[2],elemento[6]))
        if opcion == "SUPERFICIE KM2":
            totales_poblacion_capital.append((elemento[2], elemento[3]))
            color = "b"

        if opcion == "NUMERO DE HABITANTES VULNERABLES":
            totales_poblacion_capital.append((elemento[2], elemento[7]))
            color = "y"
        if opcion == "NUMERO DE HABITANTES CLASE MEDIA":
            color = "b"
            totales_poblacion_capital.append((elemento[2],elemento[8]))
        if opcion == "NUMERO DE HABITANTES CLASE ALTA":
            totales_poblacion_capital.append((elemento[2], elemento[9]))
            color = "black"  
        if opcion == "INDICE DE DESARROLLO HUMANO":
            totales_poblacion_capital.append((elemento[2], elemento[14]))
            color = "black"

        if opcion == "PORCENTAJE HABITANTES CON EDAD TRABAJAR":
            totales_poblacion_capital.append((elemento[2], elemento[10]))
            color = "black"

        if opcion == "PORCENTAJE HABITANTES ECONOMICAMENTE ACTIVOS":
            totales_poblacion_capital.append((elemento[2], elemento[11]))
            color = "black"

        if opcion == "PORCENTAJE HABITANTES OCUPADOS":
            totales_poblacion_capital.append((elemento[2], elemento[12]))
            color = "black"       

        if opcion == "PORCENTAJE HABITANTES DESOCUPADOS":
            totales_poblacion_capital.append((elemento[2], elemento[13]))
            color = "black"
        if opcion == "MUNICIPIOS":
            totales_poblacion_capital.append((elemento[2], elemento[1]))
            color = "black"   
 
    
    totales_poblacion_capital.pop(0)
    #al ser una lista de tuplas elimino el elemento 0 que corresponde al encabezado 
    

    #aqui convierto la lista de tuplas en un diccionario 

    for i in totales_poblacion_capital:  
       resultado_totales_caracteristica.setdefault(i[0],[]).append(i[1])


    flotante = False
    #verifico si la opcion ingresada por el usuario es en PORCENTAJES O INDICES para tratar los datos como flotantes
    if opcion.find("PORCENTAJE") !=-1 or opcion.find("INDICE") !=-1:
        flotante =True
        resultado_totales_caracteristica["Bogota"].sort(key=float)
        sub_punto_poblacion_bogota = float(resultado_totales_caracteristica["Bogota"][0])
    #en caso de que sea una opcion que no requiera flotantes los trato como enteros(numero de habitantes etc...)
    else:
        resultado_totales_caracteristica["Bogota"].sort(key=int)
        sub_punto_poblacion_bogota = int(resultado_totales_caracteristica["Bogota"][0])
    #aqui extraigo el valor de bogota cundinamarca  en la variable sub_punto_poblacion_bogota  y lo elimino del 
    #diccionario para tener un solo valor de bogota en el diccionario {capital:[valor_caracteristica]}
    
    resultado_totales_caracteristica["Bogota"].pop(0)


    
    #traigo la libreria que me da el mapa de colombia con las medidas establecidas etc....
    mapa_libreria = cf.NaturalEarthFeature(category='cultural',  
    name='admin_0_boundary_lines_land',
    scale='10m',facecolor='none')  #aqui me traigo el mapa utilizando el modulo cf de la clase feature
    stt_prv = cf.NaturalEarthFeature(category='cultural', 
    name='admin_1_states_provinces_lines',
    scale='10m',facecolor='none')
    fig = plt.figure(figsize=(90, 90))     #planteo el tamaño inicial del mapa   
    ax = fig.add_subplot(2, 5, 10, projection=ccrs.PlateCarree())
    ax.set_extent([-65, -80, -5, 13])
    ax.set_title(opcion, fontsize=15)
    ax.coastlines()
    valores_ordenamiento=[]
    
    #aqui obtengo el mayor valor del diccionario anterior bogota cundinamarca al no estar me asegure de que en la lista de la 
    #llave bogota siempre quedara el valor mayor y el menor lo almacene en la variable anterior
    #esto tambien evita que al graficar dos puntos en una misma coordenada se superpongan para esto uso lo anterior
    #asegurandome de que cuando sea bogota siempre grafique el mayor valor de la caracteristica y luego el menor
    #ya que de lo contrarior el de mayor area taparia al circulo de menor area
    for llave, valor in resultado_totales_caracteristica.items():
        for total in valor:
            if flotante:
                valores_ordenamiento.append(float(total))
            else:    
                valores_ordenamiento.append(int(total))
    valores_ordenamiento = sorted(valores_ordenamiento)  
    valor_mayor = valores_ordenamiento[-1]
    #obtengo el mayor valor de la caracteristica que estoy analizando y defino unos limites para hacer la regla de tres
    #que definen la proporcionalidad de los puntos
    # valor inicial circulo = 500   valor maximo caracteristica = valor_mayor
    #quiere decir que todos los demas puntos van a ser proporcionalmente mas pequeños que el punto con mayor tamaño 
    #es decir el dato mas grande de la caracteristica que estoy analizando
    


    for llave, valor in ciudades.items():
        for coordenadas in range(len(valor)-1):
            if llave in resultado_totales_caracteristica:
                    if flotante:
                        #si es flotante entonces los resultados los trato en ese tipo de dato
                        total_datos = float(resultado_totales_caracteristica[llave][0])
                        radio_circulo = (total_datos * 500) /valor_mayor
                    else:
                        #sino como enteros
                        total_datos = int(resultado_totales_caracteristica[llave][0])
                        radio_circulo = (total_datos * 500) /valor_mayor 
            #si la llave de ciudades(la capital) no esta en mi diccionario quiere decir que no tenia datos en el csv
            #por lo tanto no la considero para el grafico 
            else:
                radio_circulo = 0
            x = valor[coordenadas]
            y = valor[coordenadas +1]
            #aqui tomo las coordenadas de las 24 capitales recordar que existen dos capitales  Bogota en el csv 
            #por lo cual una estara contenidad dentro del mismo punto 
            ax.scatter(x,y,
                   transform=ccrs.PlateCarree(),
                   marker='o',s=radio_circulo, c='r')
    #aqui grafico siempre unicamente el valor de bogota que tenga el menor valor de la caracteristica
    #y el otro bogota lo grafique en el ciclo anterior asegurandome de que primero se grafique el mayor valor 
    #de los dos bogota del csv y luego el mennor
    x = ciudades["Bogota"][0]
    y = ciudades["Bogota"][1]
    radio_circulo_subpunto =   (sub_punto_poblacion_bogota * 500) /valor_mayor 
    ax.scatter(x,y,
    transform=ccrs.PlateCarree(),
    marker='o',s=radio_circulo_subpunto, c=color)  
    ax.add_feature(mapa_libreria,     linewidth=0.5, edgecolor='black', zorder=0)
    ax.add_feature(stt_prv, linewidth=0.2, edgecolor='black', zorder=5)
    plt.show()
    print("SUPERFICIE KM2")
    print("POBLACION")
    print("NUMERO DE HABITANTES POBRES")
    print("NUMERO DE HABITANTES VULNERABLES")
    print("NUMERO DE HABITANTES CLASE MEDIA")
    print("NUMERO DE HABITANTES CLASE ALTA")
    print("INDICE DE DESARROLLO HUMANO")
    print("PORCENTAJE HABITANTES CON EDAD TRABAJAR")
    print("PORCENTAJE HABITANTES ECONOMICAMENTE ACTIVOS")
    print("PORCENTAJE HABITANTES OCUPADOS")
    print("PORCENTAJE HABITANTES DESOCUPADOS")
    print("MUNICIPIOS")
    print("0 PARA SALIR DEL PROGRAMA")
    #aqui vuelvo a leer la opcion del usuario y cuando sea 0 termina la ejecucion del algoritmo
    resultado_totales_caracteristica.clear()  
    opcion = input("INGRESE LA VARIABLE BASE PARA GRAFICAR DE FORMA PROPORCIONAL LOS PUNTOS:"+"\n")

   

        
```
